```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Data points
R = np.array([0, 0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.4])
chord_length = np.array([0.3, 0.175, 0.125, 0.09, 0.075, 0.06, 0.051, 0.05])

# Define the negative exponential function with shifts
def shifted_negative_exponential(x, a, b, c, d):
    return a * np.exp(-b * (x - c)) + d

# Fit the data
params, covariance = curve_fit(shifted_negative_exponential, R, chord_length, p0=[1, 1, 0, 0])
a, b, c, d = params

# Generate points for plotting the fit
R_fit = np.linspace(min(R), max(R), 500)
chord_fit = shifted_negative_exponential(R_fit, a, b, c, d)

# Plot the data and the fit
plt.scatter(R, chord_length, label="Data Points", color="blue")
plt.plot(R_fit, chord_fit, label=f"Fit: $y = {a:.3f} e^{{-{b:.3f}(x - {c:.3f})}} + {d:.3f}$", color="red")
plt.xlabel("Radius (m)")
plt.ylabel("Chord Length (m)")
plt.title("Chord Length vs Radius Shifted Negative Exponential Fit")
plt.legend()
plt.grid()
plt.show()

# Output the parameters
print(f"Fitted equation: y = {a:.3f} * exp(-{b:.3f} * (x - {c:.3f})) + {d:.3f}")
```
